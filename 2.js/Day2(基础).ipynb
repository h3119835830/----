### 1. 运算符(跟Java差不多)
#### 1.1 赋值运算符
+= -= *= /= %=
#### 1.2 一元运算符
1. 例：正负号
2. 自增++和自减--(前置和后置有区别)
- 如let i = 1 console.log(i++ + 1) 结果为2，因为先计算i+1的2计入控制台，再将i变为2
- let i = 1 console.log(i++ + ++i + i) 结果为1+3+3=7
#### 1.3 比较运算符
1. ==：左右两边值是否相等 ===：左右两边是否类型和值相等
2. console.log(2 == '2') true 因为比较运算符有隐式转换，把'2'转换为2 双等号 只判断值 console.log(2 === '2') 判断 值 和 数据类型都一样才行 开发一般用全等
3. 字符串比较，是比较的字符对应的ASCII码
- 从左往右依次比较
- 如果第一位一样再比较第二位，以此类推
- 比较的少，了解即可
#### 1.4 逻辑运算符
和java相似 && || !
### 2. 语句
#### 2.1 分支语句
1. if分支语句：单分支、双分支、多分支
2. 三元运算符：条件 ？满足条件执行的代码 ：不满足条件执行的代码
- 数字补0案例：
- 需求：用户输入1个数，如果数字小于10，则前面进行补0，比如09 03等分析：利用三元运算符 补0 计算
- 写法：num = num < 10 ? 0 + num : num
3. switch语句
- 语句：switch(数据){
  case 值1:
      代码1
      break
  case 值2:
      代码2
      break
  default:
      代码n
      break
}
#### 2.2 循环语句
1. 断点调试：可以更好理解代码运行，工作时可以更快找到bug，使用浏览器打开调试界面
- 1. 按F12打开开发者工具
- 2. 点到source一栏
- 3. 选择代码文件
2. while(循环条件){
  要重复执行的代码(循环体)
}
3. 循环退出
- 1. break：退出循环
- 2. continue：结束本次循环，继续下次循环
